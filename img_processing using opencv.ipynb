```python
import cv2
# import numpy as np
import matplotlib.pyplot as plt
```

```python
frame = cv2.imread("thor.jpg")
```

```python
plt.imshow(frame)
```

```python
rgb = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
#cv2 imread just reverse the order of color like rgb = bgr
# In order to convert it back to its original form rgb we use the above code\
```

```python
plt.imshow(rgb)
```

#Face Detection

```python
classifier = cv2.CascadeClassifier("C:\\Users\\user\\Downloads\\haarcascade_frontalface_default.xml")
```

```python
faces = classifier.detectMultiScale(frame)
```

```python
face = faces[0]
```

```python
x,y,w,h = face
```

```python
face = rgb[y:y+h, x:x+w]
```

```python
plt.imshow(face)
```

```python
out = cv2.rectangle(rgb,(x,y),(x+w,y+h),(255,255,0),4)#image,xy axis, width and height,color,thickness
```

```python
plt.imshow(out)
```
